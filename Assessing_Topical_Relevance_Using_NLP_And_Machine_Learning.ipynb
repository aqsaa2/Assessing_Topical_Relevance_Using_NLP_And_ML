# Preprocessing

```python
import re
import spacy

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

def preprocess_text(raw_file, output_file):
    # Read the raw content
    with open(raw_file, 'r', encoding="utf-8") as f:
        text = f.read()

    # Lowercase the text
    text = text.lower()

    # Remove punctuation and digits
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)

    # Process the text with SpaCy
    doc = nlp(text)

    # Lemmatize and remove stopwords
    processed_text = " ".join(token.lemma_ for token in doc if not token.is_stop)

    # Save the processed text
    with open(output_file, 'w', encoding="utf-8") as f:
        f.write(processed_text)

    return processed_text

# Example usage
processed_text = preprocess_text('/content/raw_.txt', '/content/processed_page1.txt')
```

# Extracting Entities using

```python
def extract_entities(processed_file, entity_file):
    # Read the processed content
    with open(processed_file, 'r', encoding="utf-8") as f:
        text = f.read()

    # Process the text with SpaCy
    doc = nlp(text)

    # Extract named entities
    entities = [ent.text for ent in doc.ents]

    # Save the entities to a file
    with open(entity_file, 'w', encoding="utf-8") as f:
        for entity in entities:
            f.write(f"{entity}\n")

    return entities

# Example usage
entities = extract_entities('/content/processed_page1.txt', '/content/entities_page1.txt')
print(f"Extracted Entities: {entities}")
```

```python
import re
import spacy

# Load SpaCy's English model for tokenization and lemmatization
nlp = spacy.load("en_core_web_sm")

# Function to preprocess text
def preprocess_text(file_path):
    # Read the file content
    with open(file_path, 'r') as file:
        text = file.read()

    # Lowercase the text
    text = text.lower()

    # Remove punctuation, digits, and special characters
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)

    # Remove stopwords and lemmatize
    doc = nlp(text)
    processed_text = " ".join(token.lemma_ for token in doc if not token.is_stop)

    # Remove duplicates
    unique_words = set(processed_text.split())
    return " ".join(unique_words)

# Preprocess relevant and irrelevant files
relevant_text = preprocess_text('/content/relevant.txt')
irrelevant_text = preprocess_text('/content/irrelevant.txt')

# Save cleaned entities to new files
with open('/content/cleaned_relevant.txt', 'w') as f:
    f.write(relevant_text)

with open('/content/cleaned_irrelevant.txt', 'w') as f:
    f.write(irrelevant_text)
```

```python
import numpy as np
import hdbscan
from sklearn.metrics.pairwise import cosine_similarity

# Convert cosine similarity to a distance matrix and ensure dtype is float64
relevant_distance = np.array(1 - cosine_similarity(relevant_embeddings), dtype=np.float64)
irrelevant_distance = np.array(1 - cosine_similarity(irrelevant_embeddings), dtype=np.float64)

# HDBSCAN clustering
clusterer = hdbscan.HDBSCAN(min_cluster_size=3, metric="precomputed")
relevant_labels = clusterer.fit_predict(relevant_distance)
irrelevant_labels = clusterer.fit_predict(irrelevant_distance)

print("Relevant Cluster Labels:", np.unique(relevant_labels))
print("Irrelevant Cluster Labels:", np.unique(irrelevant_labels))
```

```python
import pandas as pd

def analyze_clusters(entities, labels, similarity_matrix):
    # Create a DataFrame for cluster analysis
    df = pd.DataFrame({
        'Entity': entities,
        'Cluster': labels,
        'Relevance Score': similarity_matrix.mean(axis=0)
    })

    # Signal 1: Percentage of entities in clusters
    clustered_percentage = 100 * (df['Cluster'] != -1).sum() / len(df)

    # Signal 2: Mean similarity of clusters vs noise
    cluster_mean = df[df['Cluster'] != -1]['Relevance Score'].mean()
    noise_mean = df[df['Cluster'] == -1]['Relevance Score'].mean()

    # Signal 3: Threshold-driven classification
    if clustered_percentage > 20 and cluster_mean >= 0.2:
        final_decision = "Highly Relevant"
    elif clustered_percentage <= 19 and cluster_mean < 0.2:
        final_decision = "Moderately Relevant"
    elif clustered_percentage <= 30 and noise_mean > 0.1:
        final_decision = "Low Relevance"
    else:
        final_decision = "Moderate Relevance"

    # Return results
    return {
        'Clustered Percentage': clustered_percentage,
        'Cluster Mean': cluster_mean,
        'Noise Mean': noise_mean,
        'Final Decision': final_decision,
        'DataFrame': df
    }

# Analyze relevant and irrelevant entities
relevant_analysis = analyze_clusters(relevant_entities, relevant_labels, relevant_similarity)
irrelevant_analysis = analyze_clusters(irrelevant_entities, irrelevant_labels, irrelevant_similarity)

# Output results
print("Relevant Analysis:", relevant_analysis)
print("Irrelevant Analysis:", irrelevant_analysis)

# Save results to CSV
relevant_analysis['DataFrame'].to_csv('/content/relevant_analysis.csv', index=False)
irrelevant_analysis['DataFrame'].to_csv('/content/irrelevant_analysis.csv', index=False)
```
